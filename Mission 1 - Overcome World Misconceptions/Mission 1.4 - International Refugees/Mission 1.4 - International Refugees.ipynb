**1. What share of the world population are international refugees?**

1. Upload the csv file to google colab

```python
from google.colab import files
uploaded = files.upload()
```

2. Use pandas to create a dataframe with the csv

```python
import pandas as pd
refugees_df = pd.read_csv("international_refugees.csv")
```

3. Create a variable with the sum of refugees

```python
sum_refugees = refugees_df['refugees'].sum()
print(sum_refugees)
```

4. Create a variable with the sum of population

```python
sum_pop = refugees_df['population'].sum()
print(sum_pop)
```

5. Print the answer to the question

```python
all_refugees_perc = (sum_refugees/sum_pop)*100

print("International refugees make " + str(all_refugees_perc) + "% of the world population")
```
