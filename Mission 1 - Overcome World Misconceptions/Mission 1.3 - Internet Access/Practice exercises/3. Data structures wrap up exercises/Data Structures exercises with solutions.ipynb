````python
shelter_dogs = {
    "fausto": {
        "name": "fausto",
        "age": 3,
        "owners": [],
        "injured": False
    },
    "cheirinhos": {
        "name": "cheirinhos",
        "age": 5,
        "owners": ["Tom"],
        "injured": True
    },
    "tobias": {
        "name": "tobias",
        "age": 7,
        "owners": ["Jane"],
        "injured": False
    },
    "pituscas": {
        "name": "pituscas",
        "age": 1,
        "owners": ["John", "Frederico", "Marie"],
        "injured": False
    },
    "george": {
        "name": "george",
        "age": 9,
        "owners": [],
        "injured": False
    },
    "potas": {
        "name": "potas",
        "age": 11,
        "owners": ["Susie", "Conrad", "Fran"],
        "injured": True
    }
}
````

Consider the following shelter dictionary to answer the following questions:

a) Create a variable for each dog with his associated age.

````python
# CODE GOES HERE
fausto = shelter_dogs["fausto"]["age"]
cheirinhos = shelter_dogs["cheirinhos"]["age"]
tobias = shelter_dogs["tobias"]["age"]
pituscas = shelter_dogs["pituscas"]["age"]
george = shelter_dogs["george"]["age"]
potas = shelter_dogs["potas"]["age"]
````

b) What is the average age of the dogs in the shelter? 

````python
# CODE GOES HERE
list_of_ages = [fausto, cheirinhos, tobias, pituscas, george, potas]
average_age = sum(list_of_ages) / len(list_of_ages)
````

c) What is the average number of owners that each dog has on the shelter? 

````python
# CODE GOES HERE
list_of_owners = [
    len(shelter_dogs["fausto"]["owners"]),
    len(shelter_dogs["cheirinhos"]["owners"]),
    len(shelter_dogs["tobias"]["owners"]),
    len(shelter_dogs["pituscas"]["owners"]),
    len(shelter_dogs["george"]["owners"]),
    len(shelter_dogs["potas"]["owners"])
]

average_owners = sum(list_of_owners) / len(list_of_owners)
average_owners
````

d) Create a tuple for `george`, `potas` and `pituscas`  with the name, age and the number of owners each one had on the past. 

````python
# CODE GOES HERE
george = (shelter_dogs["george"]["name"], shelter_dogs["george"]["age"], len(shelter_dogs["george"]["owners"]))
potas = (shelter_dogs["potas"]["name"], shelter_dogs["potas"]["age"], len(shelter_dogs["potas"]["owners"]))
pituscas = (shelter_dogs["pituscas"]["name"], shelter_dogs["pituscas"]["age"], len(shelter_dogs["pituscas"]["owners"]))

print("This is the tuple of George", george)
print("This is the tuple of Potas", potas)
print("This is the tuple of Pituscas", pituscas)
````
