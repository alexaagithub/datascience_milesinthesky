### Mission 2 - Chapter 1

- we are only playing once -> no need to check if the player has money
- there are only 2 possibilities (red and black);
- if the roulette lands on the player’s chosen color, the player wins double the amount of their bet. If they lose, the player forfeits the entire amount they bet.
- no need to ask for the user input
- the player starts with a balance of 1000€. At the end, you must update the balance accordingly to reflect a win or loss.

```python
import random
balance = 1000 # the player starts with a balance of €1000
bet_amount = 10 # amount to bet (it has to be lower than the initial_balance)
bet_choice = "red" # choose "red" or "black"

# use of choice() method from random library (imported above) to select a random value from a list with 'red' and 'black'
outcome = random.choice(['red', 'black'])

#there are only 2 possibilities, thus we only need if and else (no elifs)
if outcome == bet_choice: #the player made the right guess in color
  change = bet_amount #change in balance. the player won, so he wins the double amount of the bet. the change is positive.
else: #the player got the color wrong
  change = -bet_amount #the player lost, so he loses the amount of the bet. the change is negative.

balance = balance + change #the balace at the end of the game

#new if else statement to print a message to the player. Note: the change is never 0
if change>0: #the player won
  print("Congrats! You won " + str(change) + "€. Your current balance is " + str(balance) + "€.")
else: #the player lost
  print("Better luck next time. You lost " + str(abs(change)) + "€. Your current balance is " + str(balance) + "€.")
```
