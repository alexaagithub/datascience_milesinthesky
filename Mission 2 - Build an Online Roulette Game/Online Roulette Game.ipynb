Tasks:

1. Ask for the user’s available betting balance at the start of each play
2. At each play, ask for the user input on his bet amount and choice (a number from 0-36 or a color - red or black);
3. Generate a random number and associated color for each spin of the wheel;
4. Validate if the user won or not (by comparing the number and/or the color generated with the user’s bet choice);
5. Depending on the result, update the user balance; 
6. At the end of each play, ask the user if he wants to keep playing (and return to step 2) or not (and close the game, printing the user final balance).
Remember that a player can’t keep playing if their balance is negative or if the amount he is betting is higher than the actual balance.

Notes:
- No need to verify if the user input is correct (‘black’, ‘red’ or a number between 0-36).
- Consider that the only betting options are one number (from 0-36) or one color (red or black).
