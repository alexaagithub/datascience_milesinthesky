## Activity 2 - Play it 50 times

- Player should be able to play the roulette 50 times continuously. For that we can use the for loop.
- At each iteration, you should randomly generate a play outcome, which should be either "black" or "red"

```python
import random

balance = 1000
bet_amount = 10
bet_choice = "red"
```

## Activity 4 - Roulette Goes Beyond Colors

1. Start by defining two lists, one containing the red numbers and other with the black numbers <br>  and 
not forgetting green

2. When the roulette stops, we get a number, code a way to get the color of the number stored into a variable

3. The game engine should be: 
- **betting colors:** 

  **win**: bet amount is added to the wallet | **loss**: bet amount is taken from the wallet
- **betting numbers:** 

  **win**: 36 times the bet amount is added to the wallet | **loss**: bet amount is taken from the wallet

4. Update the logic to consider colors and numbers

5. Test your code with "red", "black" or an integer between 0-36 to confirm the user can bet colors or numbers

Red numbers: 32,19,21,25,34,27,36,30,23,5,16,1,14,9,18,7,12,3

Black numbers: 15,4,2,17,6,13,11,8,10,24,33,20,31,22,29,28,35,26

```python
# Starting point
import random

player_wallet = 1000
plays = list(range(0,5))

# List of possibilites
list_of_green = [0]
list_of_red = [32,19,21,25,34,27,36,30,23,5,16,1,14,9,18,7,12,3]
list_of_black = [15,4,2,17,6,13,11,8,10,24,33,20,31,22,29,28,35,26]

all_possible_numbers = list_of_green + list_of_red + list_of_black

# Roulette Play Engine
for play in plays:
  bet_amount = 10
  bet_choice = 15
  number = random.choice(range(0,37))

  # What's the color of the number
  if number in list_of_green:
    number_color = 'green'
  elif number in list_of_black:
    number_color = 'black'
  elif number in list_of_red:
    number_color = 'red'
  else:
    print(f"something went wrong - {number}")

  # Validating win or loss

  # colors
  if bet_choice in ['red', 'black']:
    if bet_choice == number_color:
      result = 'win'
      player_wallet += bet_amount
    else:
      result = 'loss'
      player_wallet -= bet_amount

  # number
  elif (int(bet_choice) in all_possible_numbers):
    if int(bet_choice) == number:
      result = 'win'
      player_wallet += 36 * bet_amount
    else:
      result = 'loss'
      player_wallet -= bet_amount
  else:
    print(f'something went wrong - validating win loss')

  print(f"{result} - your bet was {bet_choice}. Roulette got the number {number}, color {number_color}. Your wallet is now {player_wallet}")  
```
