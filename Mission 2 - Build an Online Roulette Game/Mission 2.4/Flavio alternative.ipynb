```python
import random
from random import randrange

balance = 1000
bet_amount = 10


red_numbers = [32, 19, 21, 25, 34, 27, 36, 30, 23, 5, 16, 1, 14, 9, 18, 7, 12, 3]
green = [0]
black_numbers = [15, 4, 2, 17, 6, 13, 11, 8, 10, 24, 33, 20, 31, 22, 29, 28, 35, 26]

res = {}
for key in red_numbers:
    res[key] = 'red'
for key in black_numbers:
    res[key] = 'black'
res[0] = 'green'

def sort(color,number):
    global balance
    srv_n_c = random.choice(list(res.items()))
    #color
    if color == srv_n_c[1]:
        balance += bet_amount
        return ('You Won, The color sorted is {}, your new balance is {} -- {}'.format(srv_n_c[1],balance,srv_n_c))
        return ()
    else:
        balance -= bet_amount
        return ('You Lost, The color sorted is {}, your new balance is {} -- {}'.format(srv_n_c[1],balance,srv_n_c))
        return (print(srv_n_c))
    # number
    if number == srv_n_c[0]:
        balance += 36*bet_amount
        return ('You Won, The number sorted is {}, your new balance is {} -- {}'.format(srv_n_c[1], balance, srv_n_c))        
    else:
        balance -= bet_amount
        return ('You Lost, The number sorted is {}, your new balance is {} -- {}'.format(srv_n_c[1], balance, srv_n_c))        

#for x in range(1, 5):
user_input=''
user_input = input ('Chose [c] for Color or [n] for Number: ')
while user_input.lower() != 'c' or 'n':
        if user_input == 'c':
            input_color = input('Choose [R]ed, [B]lack or [G]reen or [E]xit:').lower()
            if input_color == 'r':
                b = sort('red','')
                print(b)
                break
            elif input_color == 'b':
                b = sort('black','')
                print(b)
                break
            elif input_color == 'g':
                b = sort('green','')
                print(b)
                break
            else:
                break
        elif user_input == 'n':
            input_num = int(input('Choose one number between 0 and 36: '))
            if (0 <= input_num <= 36):
                c=sort('',input_num)
                print(c)
                break
        else:
            print('Wrong Number')
            break
```
